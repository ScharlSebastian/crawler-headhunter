```python
process = CrawlerProcess()
process.crawl(GoodReadsSpider)
process.start()
```

# X-Path selection

### dom durchgehen mit "/"
- /html/body/div[1]


### get element with abbrate Variant "@"
- /html/body/div[1]/h1[@id='title']
    - xpath wählt nur die h1(es) aus, deren id = title ist
- //h4[@class='sub-title']
    - um nicht den fullpath von xpath anzugeben wird "//" verwendet, um alle h4 mit dieser attributsauswahl zu finden

### Axis
- /descendant-or-self::node()
    - "/" gibt an, dass man im Root startet
    - "descendant-or-self" gibt an in welche Richtung man sich von dem Attribute das man festlegt geht
        - descendant = nach unten
        - ancestor = nach oben
    - "node()" bedeutet, wähle alle nodes aus: sei es command, html-element, js etc.
        - wähle alle kinder und kindeskinder unterhalb der node aus, egal was es für ein element ist
        - folglich wird hier alles ausgewählt
- /descendant-or-self::node()/child::h4[attribute::class='sub-title']
    - für das whofinance Beispiel:
        - /descendant-or-self::node()/child::a[attribute::class='btn btn-block btn-success btn-to-profile']
- //li/parent::node()
    - hier wird das Elternelement von li ausgewählt -> ul
- //li/ancestor::node()
    - hierbei werden alle Eltern und Großelternelement von li ausgewählt
- //li/ancestor-or-self::node()
    - mit "-or-self" werden die eigenen li Element auch noch ausgewählt
    
### Predecicates [ ]
- Hier können allemöglichen Mathematischen oder If Conditions eingebaut werden

#### position()
- //li[position() = 3 or position() = 4]
- oder //li[position() = last()-1 or position() = last()]
- oder //li[position() >= 3]

#### node and text
- //*[contains(@class, 'title')]
    - alles, was die Klasse "title" hat
    - contains ist Case-sensitive
- //*[starts-with(text(), 'Hallo')]
